```python
import numpy as np
```

```python
normal_data=np.random.normal(loc=50,scale=3,size=100)
normal_data
```

```python
o=np.array([1000,1500,2000])
nm=np.concatenate([normal_data,o])
nm
```

```python
n_=np.mean(normal_data)
o_=np.mean(nm)
print(n_)
print(o_)
```

```python
import pandas as pd
```

```python
s=pd.DataFrame({"normal_data":normal_data})
s
```

```python
s1=pd.DataFrame({"outliers_data":nm})
s1
```

```python
s.kurtosis()
```

```python
s1.kurtosis()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)

sns.histplot(normal_data,kde=True)
plt.title("normal_data")
plt.subplot(1,2,2)

sns.histplot(nm,kde=True)
plt.title("outliers_data")
plt.show()
```
